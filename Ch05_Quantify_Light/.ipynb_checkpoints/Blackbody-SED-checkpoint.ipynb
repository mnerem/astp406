# Blackbody SED plots

```python
import numpy as np
from scipy.constants import *
import plotly.express as px
```

```python
def Rayleigh_Jeans(位,T):
    # 位 in nanomters
    return 8*pi*k*T/(位*1e-9)**4
```

```python
em_spectrum=np.arange(200,700) # visible band = [380,700], 位=200nm is UV
T=[300,500,1000,2000]
```

```python
fig = px.line(x=em_spectrum,y=Rayleigh_Jeans(em_spectrum,0))
for tt in T:
    fig.add_scatter(x=em_spectrum,y=Rayleigh_Jeans(em_spectrum,tt))
```

```python
fig.show()
```
